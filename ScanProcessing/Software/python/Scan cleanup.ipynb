```python
def binaryArray(xygrid,value, bellow_threshold):
    blank = zeros(xygrid.shape)
    for i in range(0,xygrid.shape[0]):
        for j in range(0,xygrid.shape[1]):
            if (value> xygrid[i,j]) and (bellow_threshold==True):
                blank[i,j]=1
            elif(value< xygrid[i,j]) and (bellow_threshold==False):
                blank[i,j]=1
    return blank
    
```

```python
def hough_transform(img_bin, theta_res=1, rho_res=1):
  nR,nC = img_bin.shape
  theta = linspace(-90.0, 0.0, ceil(90.0/theta_res) + 1.0)
  theta = concatenate((theta, -theta[len(theta)-2::-1]))
 
  D = sqrt((nR - 1)**2 + (nC - 1)**2)
  q = ceil(D/rho_res)
  nrho = 2*q + 1
  rho = linspace(-q*rho_res, q*rho_res, nrho)
  H = zeros((len(rho), len(theta)))
  for rowIdx in range(nR):
    for colIdx in range(nC):
      if img_bin[rowIdx, colIdx]:
        for thIdx in range(len(theta)):
          rhoVal = colIdx*cos(theta[thIdx]*pi/180.0) + \
              rowIdx*sin(theta[thIdx]*pi/180)
          rhoIdx = nonzero(abs(rho-rhoVal) == min(abs(rho-rhoVal)))[0]
          H[rhoIdx[0], thIdx] += 1
  return rho, theta, H
```

```python
def maxpoints(H,percent):
    maxp = amax(H)
    points=[]
    for i in range(0,H.shape[0]):
        for j in range(0,H.shape[1]):
            if percent*maxp <H[i,j] :
                points.append((i,j))
    return points

print maxpoints(H,.999)
    
```

```python
def YfromRhoTheta(r,t,X):
    def y(r,t,x):
        return -cos(t)/sin(t) * x + r/sin(t)
    Y = zeros((X.shape[0],1))
    for i in range(0,X.shape[0]):
        Y[i] = y(r,t,X[i])
    return Y

X = arange(0,xy2.shape[0],1)
Y = arange(0,xy2.shape[1],1)
Y1 = YfromRhoTheta(Rho[244]+5,pi/90.0*Theta[92],Y)
Y2 = YfromRhoTheta(Rho[416]-8,pi/90.0*Theta[91],Y)

fig, ax = plt.subplots(1,1)
ax.plot(Y,Y1,'w')
ax.plot(Y,Y2,'y')
ax.imshow(xy2)
#print "C=",Rho[416]/sin(pi/90.0*Theta[91])
#print "M=",-cos(pi/90.0*Theta[91])/sin(pi/90.0*Theta[91])
#print "R=",Rho[416]
#print "Theta=",Theta[91]
#print Y1
```

```python
import cv2

def TwodToImage(tdarray):
    newArray = zeros((tdarray.shape[0],tdarray.shape[1],3),dtype=uint8)
    for i in range(0,tdarray.shape[0]):
        for j in range(0,tdarray.shape[1]):
            newArray[i,j,0]=tdarray[i,j]
            newArray[i,j,1]=tdarray[i,j]
            newArray[i,j,2]=tdarray[i,j]
    return newArray

def d2plot(pltarray):
    plt.figure(figsize=(12*2, 3*2))
    plt.subplot(111)
    plt.imshow(pltarray, cmap=plt.cm.gray)
    

ba3d = TwodToImage(ba)
print ba3d.shape
bw = cv2.cvtColor(ba3d, cv2.COLOR_BGR2GRAY)
d2plot(bw)

contours, hierarchy = cv2.findContours(bw, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
maxarea=0
maxcontour = contours[0]
for contour in contours:
    a = cv2.contourArea(contour)
    if(a>maxarea):
        maxarea=a
        maxcontour=contour

plot(maxcontour[:,0,0],maxcontour[:,0,1],"r")
print maxcontour.shape
```
