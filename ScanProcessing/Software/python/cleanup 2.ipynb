```python
import csv
from scipy import *
from scipy import stats, ndimage
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.gridspec as gridspec
from pylab import *
import cv2


def readFile(file):
    scale = 1
    xygrid =[]
    with open(file,'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        i=0
        for row in reader:
            if(i==0):
                scale = float(row[0])

            elif(i>2):  
                grid_row =[]
                for j in range(0,len(row)):
                    grid_row.append(float(row[j]))
                xygrid.append(grid_row)
            
            i+=1
    return (scale,array(xygrid))


```

```python
def binaryArray(xygrid,value, bellow_threshold):
    blank = zeros(xygrid.shape)
    for i in range(0,xygrid.shape[0]):
        for j in range(0,xygrid.shape[1]):
            if (value> xygrid[i,j]) and (bellow_threshold==True):
                blank[i,j]=1
            elif(value< xygrid[i,j]) and (bellow_threshold==False):
                blank[i,j]=1
    return blank
```

```python
def hough_transform(img_bin, theta_res=1, rho_res=1):
  nR,nC = img_bin.shape
  theta = linspace(-90.0, 0.0, ceil(90.0/theta_res) + 1.0)
  theta = concatenate((theta, -theta[len(theta)-2::-1]))
 
  D = sqrt((nR - 1)**2 + (nC - 1)**2)
  q = ceil(D/rho_res)
  nrho = 2*q + 1
  rho = linspace(-q*rho_res, q*rho_res, nrho)
  H = zeros((len(rho), len(theta)))
  for rowIdx in range(nR):
    for colIdx in range(nC):
      if img_bin[rowIdx, colIdx]:
        for thIdx in range(len(theta)):
          rhoVal = colIdx*cos(theta[thIdx]*pi/180.0) + \
              rowIdx*sin(theta[thIdx]*pi/180)
          rhoIdx = nonzero(abs(rho-rhoVal) == min(abs(rho-rhoVal)))[0]
          H[rhoIdx[0], thIdx] += 1
  return rho, theta, H
```

```python
def plot2DArray(d2array):
    fig = plt.figure(figsize=(12*2, 3*2))
    ax = plt.subplot(111)
    p=plt.imshow(d2array)
    ax.set_aspect(2)
    cb = fig.colorbar(p, shrink=0.5)
```

```python
def maxpoints(H,percent):
    maxp = amax(H)
    points=[]
    for i in range(0,H.shape[0]):
        for j in range(0,H.shape[1]):
            if percent*maxp <H[i,j] :
                points.append((i,j))
    return points
```

```python
def MakeSideLines(xy2):

        ba=binaryArray(xy2,.1,True)

        Rho,Theta,H = hough_transform(ba)

        points = maxpoints(H,.92)
        
        #print points
        
        
        X = arange(0,xy2.shape[0],1)
        Y = arange(0,xy2.shape[1],1)
        Ys=[]
        Left_pxo=-10*xy2.shape[0]
        Left_pxm=-10*xy2.shape[0]
        Right_pxo=10*xy2.shape[0]
        Right_pxm=10*xy2.shape[0]
        
        
        midpoint = xy2.shape[0]/2
        
        for point in points:
            if (Theta[point[1]]>0.01):
                C = Rho[point[0]]/sin(pi/90.0*Theta[point[1]])
                M = -cos(pi/90.0*Theta[point[1]])/sin(pi/90.0*Theta[point[1]])
                
                #print ""
                #print "C: ",C
                #print "M: ",M
                pxo = (-C/M)
                pxm = (xy2.shape[1]-C)/M
                #print "Px_o: ",pxo
                #print "Px_m: ",pxm
                
                if(pxo<midpoint and pxm<midpoint):
                    if pxo>Left_pxo : Left_pxo = pxo
                    if pxm>Left_pxm : Left_pxm = pxm
                elif(pxo>midpoint and pxm>midpoint):
                    if pxo<Right_pxo : Right_pxo = pxo
                    if pxm<Right_pxm : Right_pxm = pxm
        #print "Left: (",0,",",Left_pxo,") , (",xy2.shape[1],",",Left_pxm,")"
        #print "Right: (",0,",",Right_pxo,") , (",xy2.shape[1],",",Right_pxm,")"
        
        
        def makeY (m,c,X,vertical=False):
            Y = zeros((X.shape[0],1))
            for i in range(0,X.shape[0]):
                Y[i] = m*X[i]-c*m
            return Y
        
        offset = 10
        
        
        if( (Right_pxm - Right_pxo)!=0):
            Right_m = xy2.shape[1]/(Right_pxm-Right_pxo)
            RightYs = makeY(Right_m,Right_pxo-offset,Y)
        else:
            RightYs = makeY(1000000,xy2.shape[0]-offset,Y)
        
        if( (Left_pxm - Left_pxo)!=0):
            Left_m = xy2.shape[1]/(Left_pxm-Left_pxo)
            LeftYs = makeY(Left_m,Left_pxo+offset,Y)
        else:
            LeftYs = makeY(1000000,0+offset,Y)
        
        #if (xy2.shape[0]<amin(RightYs)) or (0>amax(RightYs)):
        #    for i in range(0,xy2.shape[0]):
        #        RightYs[i] = xy2d.shape[0]-offset
    
        #if (xy2.shape[0]>amax(LeftYs)) or (xy2.shape[0]>amin(LeftYs)):
        #    for i in range(0,xy2.shape[0]):
        #        LeftYs[i] = offset
        
        return xy2, LeftYs,RightYs
```

```python
def cleanBoarders(LeftYs, RightYs, xygrid):
    #print LeftYs.shape
    #print RightYs.shape
    #print xygrid.shape
    
    for i in range(0,xygrid.shape[0]):

        for j in range(0,xygrid.shape[1]):
            if i<LeftYs[j]:
                xygrid[i,j]=0
                
            elif i>RightYs[j]:
                xygrid[i,j]=0
    return xygrid
```

```python
def TwodToImage(tdarray):
    newArray = zeros((tdarray.shape[0],tdarray.shape[1],3),dtype=uint8)
    for i in range(0,tdarray.shape[0]):
        for j in range(0,tdarray.shape[1]):
            newArray[i,j,0]=tdarray[i,j]
            newArray[i,j,1]=tdarray[i,j]
            newArray[i,j,2]=tdarray[i,j]
    return newArray
```
