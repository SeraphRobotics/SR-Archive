```python
#%matplotlib inline
import csv
from scipy import *
from scipy import stats, ndimage
from scipy.spatial import ConvexHull
from scipy.interpolate import interp1d,PiecewisePolynomial
from scipy.misc import comb

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.gridspec as gridspec
from pylab import *

import cv2

from xml.etree.ElementTree import ElementTree, Element 
import xml.etree.ElementTree as etree

def readFile(file):
    scale = 1
    xygrid =[]
    with open(file,'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        i=0
        for row in reader:
            if(i==0):
                scale = float(row[0])

            elif(i>2):  
                grid_row =[]
                for j in range(0,len(row)):
                    grid_row.append(float(row[j]))
                xygrid.append(grid_row)
            
            i+=1
    return (scale,array(xygrid))


def plotGrid(xygrid, e,a,ar):
    row_j = 0
    len_i = xygrid.shape[0]
    len_j = xygrid.shape[1]
    lin_i = arange(0,len_i,1)#linspace(0,len_i,num = len_i)
    lin_j = arange(0,len_j,1)#linspace(0,len_j,num = len_j)
    X,Y = meshgrid(lin_j,lin_i)

    #print len_j
    #print lin_j
    #print len(xygrid[20])
    Z = xygrid
    Z_row = xygrid[row_j,:]
    #print Z.shape
    #print X.shape
    #print Y.shape

    #figure()
    #plot(lin_j,Z_row,'r')
    #show()

    fig = plt.figure(figsize=(14*3,6*3))

    ax = fig.add_subplot(1, 2, 2, projection='3d')
    p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
    cb = fig.colorbar(p, shrink=0.5)

    #ax.set_aspect('equal')
    ax.view_init(elev=e, azim=a)
    ax.set_aspect(ar)
    plt.show()

def binaryArray(xygrid,value, bellow_threshold):
    blank = zeros(xygrid.shape)
    for i in range(0,xygrid.shape[0]):
        for j in range(0,xygrid.shape[1]):
            if (value> xygrid[i,j]) and (bellow_threshold==True):
                blank[i,j]=1
            elif(value< xygrid[i,j]) and (bellow_threshold==False):
                blank[i,j]=1
    return blank

def hough_transform(img_bin, theta_res=1, rho_res=1):
  nR,nC = img_bin.shape
  theta = linspace(-90.0, 0.0, ceil(90.0/theta_res) + 1.0)
  theta = concatenate((theta, -theta[len(theta)-2::-1]))
 
  D = sqrt((nR - 1)**2 + (nC - 1)**2)
  q = ceil(D/rho_res)
  nrho = 2*q + 1
  rho = linspace(-q*rho_res, q*rho_res, nrho)
  H = zeros((len(rho), len(theta)))
  for rowIdx in range(nR):
    for colIdx in range(nC):
      if img_bin[rowIdx, colIdx]:
        for thIdx in range(len(theta)):
          rhoVal = colIdx*cos(theta[thIdx]*pi/180.0) + \
              rowIdx*sin(theta[thIdx]*pi/180)
          rhoIdx = nonzero(abs(rho-rhoVal) == min(abs(rho-rhoVal)))[0]
          H[rhoIdx[0], thIdx] += 1
  return rho, theta, H

def plot2DArray(d2array):
    fig = plt.figure(figsize=(12*2, 3*2))
    ax = plt.subplot(111)
    p=plt.imshow(d2array)
    ax.set_aspect(2)
    cb = fig.colorbar(p, shrink=0.5)
    
def maxpoints(H,percent):
    maxp = amax(H)
    points=[]
    for i in range(0,H.shape[0]):
        for j in range(0,H.shape[1]):
            if percent*maxp <H[i,j] :
                points.append((i,j))
    return points

def TwodToImage(tdarray):
    newArray = zeros((tdarray.shape[0],tdarray.shape[1],3),dtype=uint8)
    for i in range(0,tdarray.shape[0]):
        for j in range(0,tdarray.shape[1]):
            newArray[i,j,0]=tdarray[i,j]
            newArray[i,j,1]=tdarray[i,j]
            newArray[i,j,2]=tdarray[i,j]
    return newArray

def MakeSideLines2(xy2):

        ba=binaryArray(xy2,.1,True)

        Rho,Theta,H = hough_transform(ba)

        #plot2DArray(H)
        
        
        points = maxpoints(H,.92)
        
        #print points
        
        
        X = arange(0,xy2.shape[0],1)
        Y = arange(0,xy2.shape[1],1)
        Ys=[]
        Left_pxo=-10*xy2.shape[0]
        Left_pxm=-10*xy2.shape[0]
        Right_pxo=10*xy2.shape[0]
        Right_pxm=10*xy2.shape[0]
        
        
        midpoint = xy2.shape[0]/2
        
        for point in points:
            if (Theta[point[1]]>0.01):
                C = Rho[point[0]]/sin(pi/90.0*Theta[point[1]])
                M = -cos(pi/90.0*Theta[point[1]])/sin(pi/90.0*Theta[point[1]])
                
                #print ""
                #print "C: ",C
                #print "M: ",M
                pxo = (-C/M)
                pxm = (xy2.shape[1]-C)/M
                #print "Px_o: ",pxo
                #print "Px_m: ",pxm
                
                if(pxo<midpoint and pxm<midpoint):
                    if pxo>Left_pxo : Left_pxo = pxo
                    if pxm>Left_pxm : Left_pxm = pxm
                elif(pxo>midpoint and pxm>midpoint):
                    if pxo<Right_pxo : Right_pxo = pxo
                    if pxm<Right_pxm : Right_pxm = pxm
        #print "Left: (",0,",",Left_pxo,") , (",xy2.shape[1],",",Left_pxm,")"
        #print "Right: (",0,",",Right_pxo,") , (",xy2.shape[1],",",Right_pxm,")"
        
        
        def makeY (m,c,X,vertical=False):
            Y = zeros((X.shape[0],1))
            for i in range(0,X.shape[0]):
                Y[i] = m*X[i]-c*m
            return Y
        
        offset = 10
        
        
        if( (Right_pxm - Right_pxo)!=0):
            Right_m = xy2.shape[1]/(Right_pxm-Right_pxo)
            RightYs = makeY(Right_m,Right_pxo-offset,Y)
        else:
            print "right pxm error"
            RightYs = zeros(Y.shape,float)
            for j in range(0,xy2.shape[1]):
                RightYs[j]=xy2.shape[1]-20

        
        if( (Left_pxm - Left_pxo)!=0):
            Left_m = xy2.shape[1]/(Left_pxm-Left_pxo)
            LeftYs = makeY(Left_m,Left_pxo+offset,Y)
        else:
            print "left pxm error"
            LeftYs = zeros(Y.shape,float)
            for j in range(0,offset*2):
                LeftYs[j]=xy2.shape[1]
            for j in range(xy2.shape[1]-offset*2,xy2.shape[1]):
                LeftYs[j]=xy2.shape[1]
        
        print xy2.shape
        print (amin(RightYs),amax(RightYs))
        print (amin(LeftYs),amax(LeftYs))
        
        #if (xy2.shape[1]<amin(RightYs)) or (0>amax(RightYs)):
        #    print "error right"
        #    for i in range(0,xy2.shape[0]):
        #        RightYs[i] = offset
    
        if (0<LeftYs[LeftYs.shape[0]/2]) : #(xy2.shape[1]>LeftYs)) or (xy2.shape[1]/2 >amin(LeftYs)
            print "error left"
            LeftYs = zeros(Y.shape,float)
            for j in range(0,offset*2):
                LeftYs[j]=xy2.shape[1]
            for j in range(xy2.shape[1]-offset*2,xy2.shape[1]):
                LeftYs[j]=xy2.shape[1]
        
        return xy2, LeftYs,RightYs

def cleanBoarders(LeftYs, RightYs, xygrid):
    #print LeftYs.shape
    #print RightYs.shape
    #print xygrid.shape
    
    for i in range(0,xygrid.shape[0]):

        for j in range(0,xygrid.shape[1]):
            if i<LeftYs[j]:
                xygrid[i,j]=0
                
            elif i>RightYs[j]:
                xygrid[i,j]=0
    return xygrid
    
    
def interpolatedPoints(pts):
    xs=[]
    ys=[]
    yis = zeros((4,2))
    xnew = np.linspace(pts[0][0], pts[-1][0], 400)
    i=0
    for pt in pts:
        xs.append(pt[0])
        ys.append(pt[1])
        yis[i][0]=pt[1]
        i+=1
    #     plt.plot(pt[0],pt[1],"go")
    #f = interp1d(xs, ys)
    #f2 = interp1d(xs, ys, kind=3)
    
    yis[0][1]=2.0/(xs[1]-xs[0])*(yis[1][0]-yis[0][0])
    
    yis[1][1]=0.15/(xs[1]-xs[0])*(yis[1][0]-yis[0][0]) #0.05
    
    yis[2][1]=0.15/(xs[3]-xs[2])*(yis[3][0]-yis[2][0])

    yis[3][1]=2.0/(xs[3]-xs[2])*(yis[3][0]-yis[2][0])

    
    f3 = interp1d(xs, ys, kind=3)#PiecewisePolynomial(xs,yis,orders=2)
    #plt.plot(xs,ys,'go',xnew,f(xnew),'-', xnew, f2(xnew),'--',xnew, f3(xnew),'r-')
    #plt.plot(xs,ys,'go',xnew, f3(xnew),'r-')
    #plt.plot(xnew, f3(xnew),'r-')
    f3_y = f3(xnew)
    return xnew,f3_y


def bernstein_poly(i, n, t):
    """
     The Bernstein polynomial of n, i as a function of t
    """

    return comb(n, i) * ( t**(n-i) ) * (1 - t)**i

def bezier_curve(points, nTimes=1000):
    """
       Given a set of control points, return the
       bezier curve defined by the control points.

       points should be a list of lists, or list of tuples
       such as [ [1,1], 
                 [2,3], 
                 [4,5], ..[Xn, Yn] ]
        nTimes is the number of time steps, defaults to 1000

        See http://processingjs.nihongoresources.com/bezierinfo/
    """

    nPoints = len(points)
    xPoints = np.array([p[0] for p in points])
    yPoints = np.array([p[1] for p in points])

    t = np.linspace(0.0, 1.0, nTimes)

    polynomial_array = np.array([ bernstein_poly(i, nPoints-1, t) for i in range(0, nPoints)   ])

    xvals = np.dot(xPoints, polynomial_array)
    yvals = np.dot(yPoints, polynomial_array)

    return xvals, yvals

def indent(elem, level=0):
    # Helper function that fixes the indentation scheme of a given Element object and all of its subelements
    # Modified from: http://infix.se/2007/02/06/gentlemen-indent-your-xml
    i = "\n" + level*"  "
    if len(elem):
        if not elem.text or not elem.text.strip():
            elem.text = i + "  "
        for e in elem:
            indent(e, level+1)
            if not e.tail or not e.tail.strip():
                e.tail = i + "  "
        if not e.tail or not e.tail.strip():
            e.tail = i
    else:
        if level and (not elem.tail or not elem.tail.strip()):
            elem.tail = i

def writeTree(output_file, tree):
    f = open(output_file, 'w')
    f.write("<?xml version=\"1.0\" encoding=\"UTF-8\" standalone=\"yes\" ?> \n")
    indent(tree.getroot())
    string = etree.tostring(tree.getroot())
    f.write(string)
    f.close()    

def exportToXDFL(xloop,yloop):
    root = Element("xdfl")
    cmd = Element("commands")
    path = Element("path")
    for i in range(0,len(xloop)):
        point = Element("point")
        x = Element("x")
        x.text="%.4f"%xloop[i]
        y = Element("y")
        y.text ="%.4f"%yloop[i]
        z = Element("z")
        z.text = "0.0"
        point.append(x)
        point.append(y)
        point.append(z)
        path.append(point)
    cmd.append(path)
    root.append(cmd)
    xdfltree = ElementTree(root)
    writeTree("loop.xdfl",xdfltree)
    

def twoListToPtlist(xlist,ylist):
    pts=[]
    for i in range(0,len(xlist)):
        pt = [0,0]
        pt[0] = xlist[i]
        pt[1] = ylist[i]
        pts.append(pt)
    return pts

def intersection(pta,ptb,othera,otherb):
    ua =(otherb[0]-othera[0])*(pta[1]-othera[1]) - (otherb[1]-othera[1])*(pta[0]-othera[0])
    ub =(ptb[0]-pta[0])*(pta[1]-othera[1])-(ptb[1]-pta[1])*(pta[0]-othera[0])
    denom = (otherb[1]-othera[1])*(ptb[0]-pta[0]) - (otherb[0]-othera[0])*(ptb[1]-pta[1])
    
    
    if (denom==0):return False
    ua/=denom
    ub/=denom
    
    if (ua < 0.0 or ua > 1.0 or ub < 0.0 or ub > 1.0):return False
    pt = [0,0]
    pt[0] = pta[0]+ua*(ptb[0]-pta[0])
    pt[1] = pta[1]+ua*(ptb[1]-pta[1])
    return pt

def ptsEqual(p1,p2):
    threshold = 0.0001
    if ( abs(p1[0]-p2[0])<threshold ) or ( abs(p1[1]-p2[1])<threshold ):return True
    return False

def insideLoop(loop,pt):
    previous_pt = loop[-1]
    intersects=0
    for loop_pt in loop:
        if intersection( previous_pt,loop_pt,pt,[0,10000,0]) and not ptsEqual(loop_pt,pt): intersects+=1
        previous_pt = loop_pt
    if intersects % 2 ==0: return False
    return True

def point_inside_polygon(x,y,poly):

    n = len(poly)
    inside =False

    p1x,p1y = poly[0]
    for i in range(n+1):
        p2x,p2y = poly[i % n]
        if y > min(p1y,p2y):
            if y <= max(p1y,p2y):
                if x <= max(p1x,p2x):
                    if p1y != p2y:
                        xinters = (y-p1y)*(p2x-p1x)/(p2y-p1y)+p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x,p1y = p2x,p2y

    return inside
    
def plotHull2(filename, threshold,pts):
    
    scale,xy = readFile(filename)
    plot2DArray(xy)
    
    sx = ndimage.sobel(xy,axis=0,mode='constant')
    sy = ndimage.sobel(xy,axis=0,mode="constant")
    sob = hypot(sx,sy)
    plot2DArray(sob)
    
    #xy2, Lefty, righty = MakeSideLines2(xy)
    #print Lefty
    #print righty
    
    #cleangrid = cleanBoarders(Lefty,righty,sob)
    cleangrid = sob
    
    newgrid = stats.threshold(cleangrid,threshmin=0,threshmax=200,newval=-1)
    plot2DArray(newgrid)
    left_slope_thresh = 10
    left_hieght_thresh = 55
    left = False
    right = False
    right_slope_thresh = 20
    right_hieght_thresh = 30
    
    candidatePoints=zeros(newgrid.shape)
    for i in range(0,candidatePoints.shape[0]):
        for j in range(0,candidatePoints.shape[1]):
            if ( j > (candidatePoints.shape[1]/2) ):
                right = True
                left = False
            else:
                right = False
                left = True
            
            if left and (newgrid[i,j]>left_slope_thresh) and (xy[i,j]<left_hieght_thresh):
                candidatePoints[i,j]=1
            if right and (newgrid[i,j]<right_slope_thresh) and (xy[i,j]<right_hieght_thresh):
                candidatePoints[i,j]=1
    
    plot2DArray(candidatePoints)
     

    '''
    ba=binaryArray(newgrid,threshold,False)
    plot2DArray(ba)
    
    
    open_img = ndimage.binary_opening(ba)
    close_img = ndimage.binary_closing(open_img)
    a = ndimage.morphology.binary_erosion(close_img, iterations=3)
    
    open_img = ndimage.binary_opening(ba)
    close_img = ndimage.binary_closing(open_img)

    a = ndimage.morphology.binary_erosion(close_img, iterations=1)
    '''
    ba=candidatePoints
    a = candidatePoints
    ba3d = TwodToImage(a)
    bw = cv2.cvtColor(ba3d, cv2.COLOR_BGR2GRAY)
    contours, hierarchy = cv2.findContours(bw, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    maxarea=0
    maxcontour = contours[0]
    for contour in contours:
        a = cv2.contourArea(contour)
        if(a>maxarea):
            maxarea=a
            maxcontour=contour
    
    #plot(maxcontour[:,0,0],maxcontour[:,0,1],"r")
    
    hull = ConvexHull(maxcontour[:,0,:])
    xsimplex=[]
    ysimplex=[]
    plt.figure(1)
    for simplex in hull.simplices:
        ax = plt.subplot(111)
        plt.plot(maxcontour[simplex,0,0], maxcontour[simplex,0,1], 'g-')
        plt.imshow(ba, cmap=plt.cm.gray)
        ax.set_aspect(2)
        for j in range(0,len(maxcontour[simplex,0,0])):
            xsimplex.append( maxcontour[simplex,0,0][j])
            ysimplex.append( maxcontour[simplex,0,1][j])

    
    for pt in pts:
        plt.plot(pt[0],pt[1],"go")
    [curve_pts_x,curve_pts_y] = bezier_curve(pts)#interpolatedPoints(pts)
    
    maxx = min(curve_pts_x)
    maxy = min(curve_pts_y)
    final_loop_x=[]
    final_loop_y=[]
    
    reject_x=[]
    reject_y=[]
    
    maxed = False
    for i in range(0,len(xsimplex)):
        if ysimplex[i]<maxy:
            final_loop_x.append(xsimplex[i])
            final_loop_y.append(ysimplex[i])
            
        elif ysimplex[i]>maxy and not maxed:
            maxed=True
            for j in range(0,len(curve_pts_x)):
                final_loop_x.append(curve_pts_x[j])
                final_loop_y.append(curve_pts_y[j])
        elif ysimplex[i]>maxy and maxed:
            reject_x.append(xsimplex[i])
            reject_y.append(ysimplex[i])
    
    numpts= len(final_loop_x)
    points = zeros( (numpts,2) )
    
    for i in range(0,len(final_loop_x)):
        points[i,0]=final_loop_x[i]
        points[i,1]=final_loop_y[i]
        
    #sort points by polar coordinates
    cent = [sum(x for x in final_loop_x)/numpts,sum(y for y in final_loop_y)/numpts]
    indecies = range(0,len(final_loop_x))
    indecies.sort(key = lambda i: math.atan2( (final_loop_y[i]-cent[1]) , (final_loop_x[i] - cent[0]) ) )

    xloop = [final_loop_x[i] for i in indecies]
    yloop = [final_loop_y[i] for i in indecies]
    
    
    loop = twoListToPtlist(yloop,xloop)
    
    plt.plot( xloop, yloop,"r-")

    exportToXDFL(xloop,yloop)

    shape_x=cleangrid.shape[0]
    shape_y=cleangrid.shape[1]
    
    for i in range(0,shape_x):
        for j in range(0,shape_y):
            gridpt = [i,j]
            if (xy[i,j] > 100):
                xy[i,j]=100
            if not point_inside_polygon(i,j,loop):#insideLoop(loop,gridpt):
                xy[i,j]=0
                   
    
    plt.figure(2)
    ax = plt.subplot(111)
    plt.imshow(xy)
    ax.set_aspect(2)
    #plotGrid(xy,90,90,2)
    #plotGrid(xy,45,-45,2)
    #plt.plot(curve_pts_x,curve_pts_y,"b--")
    #plt.plot(final_loop_x,final_loop_y,"r-")

    #plt.plot(points[new_hull.vertices[0],0], points[new_hull.vertices[0],1], 'ro')
    #plt.plot(reject_x,reject_y,'b-')
    #plt.plot(xsimplex,ysimplex,'g-')
    #print final_loop_x,final_loop_y
    #return xy,cleangrid,newgrid,ba,a
```
