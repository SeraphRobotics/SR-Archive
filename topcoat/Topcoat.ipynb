```python
import csv
from scipy import *
from scipy import stats, ndimage
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.gridspec as gridspec
from pylab import *


def readFile(file):
    scale = 1
    xygrid =[]
    with open(file,'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        i=0
        for row in reader:
            if(i==0):
                scale = float(row[0])

            elif(i>2):  
                grid_row =[]
                for j in range(0,len(row)):
                    grid_row.append(float(row[j]))
                xygrid.append(grid_row)
            
            i+=1
    return (scale,array(xygrid))

```

```python
def offsetZ(grid, zamount):
    for x in nditer(grid, op_flags=['readwrite']):
        x[...]= x+zamount
    return grid
```

```python
def makePtsList(grid,start,stop,pw,scalex,scaley):
    x_path_points = arange(start+pw/2,stop-1,pw)
    #print x_path_points
    pts_list =[]
    
    j=0
    
    for x in x_path_points:
        j+=1
        range_list=[]
        
        if j % 2 ==0: range_list = range(0,grid.shape[1]-1)
        else: range_list = range(grid.shape[1]-1,-1,-1)
        
        pt_list = []
        for i in range_list:
            pt = [0,0,0]
            pt[0]=x
            pt[1]=i*scaley
            
            xjminus = floor(x/scalex)
            xminus = scalex*xjminus
            zminus = grid[xjminus,i]
            
            xjplus = ceil(x/scalex)
            xplus = scalex*xjplus
            zplus = grid[xjplus,i]
            
            m = (zplus-zminus)/(xplus-xminus)
            b = zplus-m*xplus
            
            z = m*x+b
            
            pt[2]=z
            pt_list.append(pt)
        pts_list.append(pt_list)
    return pts_list

        
#pts_list = makePtsList(grid,start,stop,pw)
#print len(pts_list)
#print len(pts_list[2])
#print len(pts_list[2][2])
    
```

```python
def ptsToGCode(pts_list, speed_in_mm_p_s):
    g_str=""
    speed = 60*speed_in_mm_p_s
    
    ## open valve
    closeLine = "\nG4 P2\nM340 P0 S1650\n"
    openLine = "\nG4 P2\nM340 P0 S2100"
    g_str+=openLine
    for pts in pts_list:
        for pt in pts:
            line="\nG1 X%.3f Y%.3f Z%.3f F%.3f"%(pt[0],pt[1],pt[2],speed)
            g_str+=line
    g_str+=closeLine
    return g_str
    
```

```python
scale,grid = readFile("Box1-sample1-Left.csv")
grid = offsetZ(grid,10)
start=0
stop = grid.shape[0]*2
pw = 1.2
pts_list = makePtsList(grid,start,stop,pw,2,1)
```

```python
gcode= ptsToGCode(pts_list,10)
f = open('workfile.gcode', 'w')
f.write(gcode)
f.close()
```

```python
def makeLoop(xo,yo,r):
    thetas = arange(0,2*math.pi,math.pi/20)
    pts =[]
    for theta in thetas:
        pt=[0,0]
        pt[0]=xo+r*math.sin(theta)
        pt[1]=yo+r*math.cos(theta)
        pts.append(pt)
    return pts

```

```python
def intersection(pta,ptb,othera,otherb):
    ua =(otherb[0]-othera[0])*(pta[1]-othera[1]) - (otherb[1]-othera[1])*(pta[0]-othera[0])
    ub =(ptb[0]-pta[0])*(pta[1]-othera[1])-(ptb[1]-pta[1])*(pta[0]-othera[0])
    denom = (otherb[1]-othera[1])*(ptb[0]-pta[0]) - (otherb[0]-othera[0])*(ptb[1]-pta[1])
    
    
    if (denom==0):return False
    ua/=denom
    ub/=denom
    
    if (ua < 0.0 or ua > 1.0 or ub < 0.0 or ub > 1.0):return False
    pt = [0,0]
    pt[0] = pta[0]+ua*(ptb[0]-pta[0])
    pt[1] = pta[1]+ua*(ptb[1]-pta[1])
    return pt

#intersection([0,0,0],[0,1,0],[-1.5,1.5,0],[1.5,1.5,0])
```

```python
def ptsEqual(p1,p2):
    threshold = 0.0001
    if ( abs(p1[0]-p2[0])<threshold ) or ( abs(p1[1]-p2[1])<threshold ):return True
    return False

def insideLoop(loop,pt):
    previous_pt = loop[-1]
    intersects=0
    for loop_pt in loop:
        if intersection( previous_pt,loop_pt,pt,[10000,10000,1000]) and not ptsEqual(loop_pt,pt): intersects+=1
        previous_pt = loop_pt
    if intersects % 2 ==0: return False
    return True

#loop = makeLoop(10,0,11)
#print insideLoop(loop,[0,10,0])
#print insideLoop(loop,[2,2,12])
```
